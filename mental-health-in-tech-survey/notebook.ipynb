# 欠損値の確認を行う

数値データではない形式が多いため、欠損値の場合なぜ欠損値になっているのかを調査し、適切な値に補完するか削除するかを検討する

```python
# 各カラム毎に欠損値の数の合計を計算する
df_missing = pd.DataFrame(df_data.isnull().sum(), columns=["num of missing"])

# 全データに締める欠損データの数を割合で確認するためのデータを準備する
total_count = df_data.shape[0] # 全データ数を取得する
df_missing_rate = df_missing[(df_missing["num of missing"] > 0)] /total_count*100 # 欠損値の数を全データ数で割ることで割合を求める

print("欠損データの割合を表示。（※ 全体のデータ件数は、", total_count, "件）")
pd.concat([df_missing[(df_missing["num of missing"] > 0)], df_missing_rate], axis=1)
```

上記より、

- state: 約41%
- self_employed: 約1%
- work_interfere: 約21%
- comments: 約87%

に欠損値が見つかった。

それぞれの欠損データの理由を考える

## 欠損値「state」の確認
stateはアメリカ合衆国に住んでいる場合州を指定するために用いられる。<br>
そのため、アメリカ以外の国に住んでいる場合は、欠損値となる場合が考えられるので、それを確かめる

```python
# Stateが欠損値の場合の、国名を一覧で表示する
print("Stateが欠損値の場合のデータを確認する")
display(df_data[df_data["state"].isnull()].head())

## Stateが欠損値の数を国別に表示する
print("Stateが欠損値の場合のデータを確認する")
display(df_data[df_data["state"].isnull()]["Country"].value_counts())
```

上記より「アメリカ以外の国に住んでいる場合は、欠損値となる」ことが分かる。<br>
この場合欠損値という扱いではないので、「no-data」という文字を登録しておくことにし、<br>
アメリカに住んでいるかどうかを新しい説明変数で持たせておくようにする。（何かに使えるかもしれないので...）

```python
df_data_c = df_data.fillna({"state":"no-data"})
df_data_c["isUSA"] = (df_data_c["state"] != "no-data")*1
df_data_c[["Country", "state", "isUSA"]].head()
```

## 欠損値「self_employed」の確認
self_employedは、自営業かどうかの判定のフラグに用いられている。<br>
欠損値の傾向を確認し、補完するかどうかを検討する

```python
# 欠測値以外のデータを確認
display(df_data_c["self_employed"].value_counts())

# YES/NOそれぞれのデータの傾向を確認
print("self_employed => Yes の場合")
display(df_data_c[df_data_c["self_employed"] == "Yes"].head())
print("self_employed => No の場合")
display(df_data_c[df_data_c["self_employed"] == "No"].head())

print("self_employed が欠損の場合: ", df_data_c["self_employed"].isnull().sum(), "件")
display(df_data_c[df_data_c["self_employed"].isnull()])
```

self_employedのデータが欠損値しているからといって、他の変数に変なデータは混ざっていなさそうな印象のため、記入漏れを疑う。<br>
自営業かどうかを従業員の数と比較するクロス集計を実施し、従業員の数をベースに自営業のデータを保管できないかを検討してみる。

```python
# 自営業と従業員規模に焦点をあててクロス集計を実施
df_no_employees = pd.crosstab(index=df_data_c["self_employed"], columns=df_data_c["no_employees"], margins=True)
display(df_no_employees)
```

自営業の場合は比較的少人数の従業員の割合であるが、少人数が必ず自営業であるとも限らない。<br>
このことから、欠損値の補完として従業員ベースに自営業であるかどうかを判定することは難しく、<br>
記入漏れのデータを未記入を表すために「no-data」という文字列で補完する方向にしてみる。<br>

```python
# self_employedの未記入データを「no-data」という文字列で補完する
df_data_c = df_data_c.fillna({"self_employed":"no-data"})
```

## 欠損値「work_interfere」の確認

精神病が仕事の妨害になると感じるかという情報である。<br>
精神病の人が積極的に回答する項目なのかなとも感じたので、念のため欠損データと精神病の有無がどう関係しているのかをチェックしておく

```python
# 欠測値以外のデータを確認
print("■ work_interfereについて、欠損値以外のデータを確認")
df_data_c["work_interfere"].value_counts()
df_work_interfere = pd.crosstab(index=df_data_c["treatment"], columns=df_data_c["work_interfere"], margins=True)
display(df_work_interfere)

print("■ work_interfereの欠損値のときの、精神病の有無についての傾向を確認")
df_work_interfere_nan = df_data_c[["treatment", "work_interfere"]][df_data_c["work_interfere"].isnull()]
df_work_interfere_nan = df_work_interfere_nan.fillna({"work_interfere":"NaN"})
pd.crosstab(index=df_work_interfere_nan["work_interfere"], columns=df_work_interfere_nan["treatment"], margins=True).drop("All").drop("All", axis=1)
```

欠損値のうち、精神病以外の人がほぼ未回答の状態ではあることが分かるが、完全に補完することはできそうにない。<br>
そのため、work_interfereについても「no-data」という文字列で補完する

```python
# work_interfere の未記入データを「no-data」という文字列で補完する
df_data_c = df_data_c.fillna({"work_interfere":"no-data"})
```

```python
## 各説明変数を確認する際に、精神病患者である「treatment_Yes」と各説明変数の値の相関係数を格納しておく
treatment_Yes={}

# 説明変数との相関係数を格納するための処理
def set_treatment_Yes_corr(df_data_c_dummies):
    corr = df_data_c_dummies.corr()
    display(corr)
    for target_column in corr["treatment_Yes"].index:
        if target_column in ["treatment_No", "treatment_Yes"]:
            continue
        # pandassに変換可能なように配列で値はセットしておく
        treatment_Yes[target_column] = [corr["treatment_Yes"][target_column]]


for column in df_data_c.columns:
    if column in ["Timestamp", "treatment", "Gender", "comments"]:
        continue

    # クロス集計とその結果を割合で表示する
    print(column)
    df_c = pd.crosstab(index=df_data_c["treatment"], columns=df_data_c[column], margins=True)
    display(df_c) 

    # 棒グラフでデータ数を確認
    print("棒グラフでデータを確認する")
    df_c = df_c.drop("All").drop("All", axis=1)
    df_c.plot.bar(stacked=True)
    plt.show()

    # 相関係数を確認
    print("ダミー変数を用いて相関係数を求める")
    df_data_c_dummies= pd.get_dummies(df_data_c[["treatment", column]])
    set_treatment_Yes_corr(df_data_c_dummies)
    
    # クロス集計を%表示する場合
    # 参考: https://teratail.com/questions/2802
#    display(df_c.apply(lambda r: r/r.sum()*100, axis=1))
```
